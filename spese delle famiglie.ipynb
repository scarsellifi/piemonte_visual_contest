```python
import os
import pandas as pd
import zipfile
```

```python
##dati importati da dati.istat.it
```

```python
z = zipfile.ZipFile(os.path.join("Average monthly expenditure.gz"))
df_dati = pd.read_csv(z.open("DCCV_SPEMMFAM_Data_a15f683b-48d9-4508-84f0-17ef60e88c1f.csv"), sep=",", encoding = "cp1252")
```

```python
piemonte = df_dati[df_dati["Territorio"] == "Piemonte"]
piemonte = piemonte[["Anno", "Gruppo di spesa", "0"]]

italia = df_dati[df_dati["Territorio"] == "Italia"]
italia = italia[["Anno", "Gruppo di spesa", "0"]]

alimentari_su_totale = [u'alimentari e bevande',
       u'non alimentari', u'totale']
alimentari_dettaglio = [u'pane e cereali', u'carne', u'pesce', u'latte, formaggi e uova',
       u'oli e grassi', u'patate, frutta e ortaggi',
       u'zucchero, caff\xe8 e drogheria', u'bevande', u'alimentari e bevande']
```

```python
piemonte = piemonte.pivot(index='Anno', columns="Gruppo di spesa", values="0")[alimentari]
italia = italia.pivot(index='Anno', columns="Gruppo di spesa", values="0")[alimentari]
```

```python
##Piemonte spesa media delle famiglie alimentare e non. 
```

```python
piemonte[alimentari_su_totale]
```

```python
## Italia spesa media delle famiglie alimentare e non. 
```

```python
italia[alimentari_su_totale]
```

```python
##Piemonte spesa media delle famiglie, dettaglio alimentare 
```

```python
piemonte[alimentari_dettaglio]
```

```python
italia[alimentari_dettaglio]
```

```python
##differenza nei consumi fra Piemonte e Italia nel 2013
```

```python
(piemonte[alimentari_dettaglio] / italia[alimentari_dettaglio] * 100 - 100).loc[2013]
```

```python
##composizione della spesa alimentare in Piemonte nel 2013
```

```python
piemonte_composizione = (piemonte[alimentari_dettaglio].divide(piemonte["alimentari e bevande"], axis = 0) * 100).loc[2013]
```

```python
piemonte_composizione
```

```python
italia_composizione = (italia[alimentari_dettaglio].divide(italia["alimentari e bevande"], axis = 0) * 100).loc[2013]
```

```python
##composizione della spesa alimentare in Italia nel 2013
```

```python
italia_composizione
```

```python
##differenze fra composizione Piemontese e italiana in termini percentuali
```

```python
piemonte_composizione / italia_composizione * 100 - 100
```
